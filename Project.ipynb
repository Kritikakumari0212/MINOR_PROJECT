Creating a Dataframe

```python
data = pd.read_csv('datasets/energy_dataset.csv', delimiter=',', nrows = None)
data.dataframeName = 'energy_dataset.csv'
nRow, nCol = data.shape
print(f'There are {nRow} rows and {nCol} columns')
```

```python
data.tail(5)
```

# Time Series Plot

```python
# Assuming "time" column contains time stamps (you may need to adjust this based on your actual data)
data['time'] = pd.to_datetime(data['time'])

# Choose the energy generation source (fossil gas)
energy_source = "generation fossil gas"

# Group the data by time and calculate the total generation for the chosen energy source
grouped_data = data.groupby('time')[energy_source].sum()

# Create a column chart to visualize the generation of the chosen energy source over time
plt.figure(figsize=(12, 6))
plt.bar(grouped_data.index, grouped_data.values, width=0.5, color='skyblue')
plt.xlabel('Time')
plt.ylabel('Generation (MW)')
plt.title(f'{energy_source} Generation Over Time')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Assuming "time" column contains time stamps (you may need to adjust this based on your actual data)
data['time'] = pd.to_datetime(data['time'],utc=True)

# Choose an energy generation source (e.g., fossil oil) and plot its generation over time
energy_source = "generation fossil oil"
plt.figure(figsize=(12, 6))
plt.plot(data['time'], data[energy_source])
plt.xlabel('Time')
plt.ylabel('Generation (MW)')
plt.title(f'{energy_source} Generation Over Time')
plt.show()
```

# Pie Chart:



```python


# Assuming "time" column contains time stamps (you may need to adjust this based on your actual data)
data['time'] = pd.to_datetime(data['time'])

# Columns for fossil fuels
fossil_fuels = ["generation fossil brown coal/lignite", "generation fossil coal-derived gas",
                "generation fossil gas"]

# Calculate the total generation for each fossil fuel source
total_generation_fossil_fuels = data[fossil_fuels].sum()

# Create a pie chart to compare the total generation from different fossil fuel sources
labels = [col.split()[-2] for col in fossil_fuels]
sizes = total_generation_fossil_fuels.values
colors = ['gold', 'lightcoral', 'skyblue']

plt.figure(figsize=(5, 5))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Proportion of Fossil Fuel Generation')
plt.axis('equal')
plt.show()
```

# Bar Chart:


```python
# Assuming "time" column contains time stamps (you may need to adjust this based on your actual data)
data['time'] = pd.to_datetime(data['time'])

# Calculate the total forecasted and actual load for the entire dataset
total_forecasted_load = data['total load forecast'].sum()
total_actual_load = data['total load actual'].sum()

# Create a bar chart to compare the proportion of forecasted and actual load
labels = ['Forecasted Load', 'Actual Load']
sizes = [total_forecasted_load, total_actual_load]
colors = ['skyblue', 'lightcoral']

plt.figure(figsize=(8, 6))
plt.bar(labels, sizes, color=colors, alpha=0.7)
plt.xlabel('Load Type')
plt.ylabel('Total Load (MW)')
plt.title('Proportion of Forecasted Load vs. Actual Load')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Assuming "time" column contains time stamps (you may need to adjust this based on your actual data)
data['time'] = pd.to_datetime(data['time'])

# Columns for fossil fuels
fossil_fuels = ["generation fossil hard coal", "generation fossil oil",
                "generation fossil oil shale", "generation fossil peat"]

# Group the data by time and calculate the total generation for each fossil fuel source
grouped_data = data.groupby('time')[fossil_fuels].sum()

# Flatten the grouped data into a single array for histogram
total_generation = grouped_data.values.flatten()

# Create a histogram to visualize the distribution of total generation of fossil fuels over time
plt.figure(figsize=(12, 6))
plt.hist(total_generation, bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Total Generation (MW)')
plt.ylabel('Frequency')
plt.title('Histogram: Total Fossil Fuel Generation Over Time')
plt.grid(True)
plt.show()
```

```python
data['time'] = pd.to_datetime(data['time'], utc=True)

# Choose the energy generation source (fossil hard coal)
energy_source = "generation fossil hard coal"

# Group the data by day and calculate the total generation for the chosen energy source
grouped_data = data.groupby(data['time'].dt.date)[energy_source].sum()

# Create a bar chart to visualize the generation of the chosen energy source over time
plt.figure(figsize=(10, 6))
plt.bar(grouped_data.index, grouped_data.values, width=1, color='skyblue')
plt.xlabel('Time')
plt.ylabel('Generation (MW)')
plt.title(f'{energy_source} Generation Over Time')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
